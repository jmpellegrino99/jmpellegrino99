```python
import numpy as np

# Define the function f(x) = x^2
def f(x):
    return np.sin(x)

# Define the range
range_ = [0, 90]

# Function to find the maximum value of f(x) within the range
def max_():
    res = f(range_[0])  # Initialize res with f(range_[0])
    for i in np.arange(range_[0], range_[1], (range_[1] - range_[0]) / 10000):
        if f(i) > res:
            res = f(i)
    return res

# Function to find the minimum value of f(x) within the range
def min_():
    res = f(range_[0])  # Initialize res with f(range_[0])
    for i in np.arange(range_[0], range_[1], (range_[1] - range_[0]) / 10000):
        if f(i) < res:
            res = f(i)
    return res

# Calculate the volume under the curve
volume = (range_[1] - range_[0]) * (max_() - min_())

# Monte Carlo simulation parameters
nsim = 10**4
counter = 0

# Perform Monte Carlo simulation
for i in range(nsim):
    x1 = np.random.uniform(range_[0], range_[1])
    y1 = np.random.uniform(min_(), max_())
    if y1 < f(x1):
        counter += 1

# Estimate the volume under the curve
estimated_volume = counter / nsim * volume

print("Estimated Volume under the Curve:", estimated_volume)
```
