```python
# Linear modeling:;

import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd

# Example data
df = pd.DataFrame({
    "y": [1, 2, 3, 4, 5],
    "x1": [2, 1, 4, 3, 5],
    "x2": [5, 3, 6, 2, 4]
})

# Run regression (similar to Stata's `reg y x1 x2`)
model = smf.ols("y ~ x1 + x2", data=df).fit()

# print(model.summary())
# print(model.params)

model.params[1:]
```
