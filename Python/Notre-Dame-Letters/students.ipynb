```python
import pandas as pd

pd.set_option('display.max_columns',None)
pd.set_option('display.width',None)

df = pd.read_csv('cleaneddata.csv')

df.fillna('', inplace=True)

df['Student_Number'] = df['Student_Number'].astype(str).str.replace(',','')
df['ID'] = df['ID'].astype(str).str.replace(',','')
df['Zip'] = df['Zip'].astype(str).str.replace(',','')
df['ClassOf'] = df['ClassOf'].astype(str).str.replace(',','')
df['Index'] = df['Index'].astype(str).str.replace(',','')

df['Amount1'] = df['Amount1'].apply(lambda x: '$' + x if x != '' else x)
df['Amount2'] = df['Amount2'].apply(lambda x: '$' + x if x != '' else x)
df['Amount3'] = df['Amount3'].apply(lambda x: '$' + x if x != '' else x)
df['Amount4'] = df['Amount4'].apply(lambda x: '$' + x if x != '' else x)

df['Total_Aid'] = '$' + df['Total_Aid'].astype(str)
df['Tuition'] = '$' + df['Tuition'].astype(str)
df['Net_Amount'] = '$' + df['Net_Amount'].astype(str)

df.to_csv('carve.csv', na_rep='', index=False)

```
